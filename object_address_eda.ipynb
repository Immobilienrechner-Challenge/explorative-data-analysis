```python
# create an function that extract address
def extract_address(df, col):
    '''
    function to extract the coloumn address into
    street, zip_code and canton

    df: DataFrame 
    col: coloumn name (address)
    '''

    # extract the street from address coloumn until the start of the first comma
    df['street'] = df[col].str.extract(r'(.+?),')
    
    # replace the whole value in street if it starts with a number
    df['street'] = df[col].str.replace(r'^\d+\s\w+', 'NaN')

    # extract the zip code from the coloumn address and remove zip code from address
    df['zip_code'] = df[col].str.extract(r'(\d{4})')

    # extract canton from address
    df['canton'] = df[col].str.extract(r'(\w{2})$')
    
    # drop the address coloumn
    df.drop([col], axis=1, inplace=True)

    return df
    
```
