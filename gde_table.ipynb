```python
import numpy as np
import pandas as pd
from helper import ImmoHelper
from sklearn.impute import KNNImputer
```

```python
# Get Data
helper = ImmoHelper()
data = helper.process_data(return_gde = True)
data_rm = helper.process_data(return_gde = False)
```

```python
# Remove all non gde columns and non numeric data
cols = list(set(data) - set(data_rm))
data = data[cols]._get_numeric_data().copy()
data
```

```python
# Group and aggregiate data based on zip code
gde_data_per_zip = data.groupby("Zip").agg(np.mean).copy()
gde_data_per_zip
```

```python
# Get missing zip codes between 1000 and 9999
all_plz = np.linspace(1000, 9999, num = 9000, dtype = int)
available_plz = np.array(gde_data_per_zip.index)
missing_plz = np.setdiff1d(all_plz, available_plz)
```

```python
# Expand Dataset with missing zip codes
empty_rows = pd.DataFrame(np.nan, index=missing_plz, columns=gde_data_per_zip.columns)
gde_data_expanded = pd.concat((gde_data_per_zip, empty_rows)).sort_index().reset_index(drop = False)
gde_data_expanded
```

```python
# Impute all NA values with K-Nearest Neighbour and divide PLZ into Groups of 100 and get dummies
imputer = KNNImputer(n_neighbors = 3, weights = "distance")
gde_data_imputed = imputer.fit_transform(gde_data_expanded)
gde_data_imputed = pd.DataFrame(gde_data_imputed, columns = gde_data_expanded.columns)
gde_data_imputed["index"] = gde_data_imputed["index"].astype(int)
gde_data_imputed.columns = gde_data_imputed.columns.str.replace("index", "PLZ")
gde_data_imputed["PLZ_Group"] = (gde_data_imputed["PLZ"] // 100) * 100
gde_data_imputed = pd.get_dummies(gde_data_imputed, columns = ["PLZ_Group"])
gde_data_imputed = gde_data_imputed.set_index("PLZ")
```
